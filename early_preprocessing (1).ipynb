## Days Since Last Match Columns

```python
display("Total Unhandle Updating new team name:")
len(df_combined[df_combined['days_since_last_match_team'].isna() & df_combined['is_train'] == True]['team']) - df_combined[df_combined['days_since_last_match_team'].isna() & df_combined['is_train'] == True]['team'].nunique()
```

```python
df_combined['days_since_last_match_team'].isna().sum()
```

```python
team_df = df_combined[['date', 'team']].rename(columns={'team': 'team_name'})
opp_df  = df_combined[['date', 'opponent']].rename(columns={'opponent': 'team_name'})

long_df = pd.concat([team_df, opp_df], axis=0)
long_df = long_df.drop_duplicates()
long_df.duplicated().sum()
```

```python
long_df = long_df.sort_values(['date', 'team_name'])
long_df['prev_date'] = long_df.groupby('team_name')['date'].shift(1)

long_df['days_since_last_match'] = (long_df['date'] - long_df['prev_date']).dt.days
long_df = long_df.drop(columns='prev_date')
```

```python
long_df
```

```python
team_days = long_df[['date', 'team_name', 'days_since_last_match']].rename(columns={
    'team_name': 'team',
    'days_since_last_match': 'days_since_last_match_team_new'
})

opp_days = long_df[['date', 'team_name', 'days_since_last_match']].rename(columns={
    'team_name': 'opponent',
    'days_since_last_match': 'days_since_last_match_opp_new'
})

df_combined = df_combined.merge(team_days, on=['date', 'team'], how='left')
df_combined = df_combined.merge(opp_days, on=['date', 'opponent'], how='left')
```

```python
df_combined['days_since_last_match_team'] = df_combined['days_since_last_match_team'].fillna(df_combined['days_since_last_match_team_new'])
df_combined['days_since_last_match_opp'] = df_combined['days_since_last_match_opp'].fillna(df_combined['days_since_last_match_opp_new'])
df_combined = df_combined.drop(columns=['days_since_last_match_team_new', 'days_since_last_match_opp_new'])
len(df_combined.columns)
```

```python
df_combined['days_since_last_match_team'].isna().sum()
```

```python
df_combined['days_since_last_match_team'] = df_combined['days_since_last_match_team'].fillna(0)
df_combined['days_since_last_match_opp'] = df_combined['days_since_last_match_opp'].fillna(0)
df_combined['first_match_team'] = (df_combined['days_since_last_match_team'] == 0).astype(int)
df_combined['first_match_opp'] = (df_combined['days_since_last_match_opp'] == 0).astype(int)
df_combined['days_since_last_match_team'].isna().sum()
```

```python
# Population And GDP based Country

# Median
# for col in ['population_team', 'gdp_per_capita_team']:
#     df_combined[col] = df_combined.groupby('team')[col].transform(lambda x: x.fillna(x.median()))

# for col in ['population_opp', 'gdp_per_capita_opp']:
#     df_combined[col] = df_combined.groupby('opponent')[col].transform(lambda x: x.fillna(x.median()))


# Interporation
df_combined = df_combined.sort_values('date')


for col in ['population_team', 'gdp_per_capita_team']:
    df_combined[col] = df_combined.groupby('team')[col].transform(lambda x: x.interpolate(method='linear', limit_direction='both'))

for col in ['population_opp', 'gdp_per_capita_opp']:
    df_combined[col] = df_combined.groupby('opponent')[col].transform(lambda x: x.interpolate(method='linear', limit_direction='both'))
```

```python
# Height Of Match Venue
df_combined['altitude_venue'] = df_combined.groupby('venue_country')['altitude_venue'].transform(lambda x: x.fillna(x.median()))
```

```python
# Miliaage
df_combined.loc[df_combined['is_home'] == 1, 'distance_travel_team'] = df_combined.loc[df_combined['is_home'] == 1, 'distance_travel_team'].fillna(0)
df_combined.loc[df_combined['is_home'] == 0, 'distance_travel_opp'] = df_combined.loc[df_combined['is_home'] == 0, 'distance_travel_opp'].fillna(0)

df_combined['distance_travel_team'] = df_combined.groupby('team')['distance_travel_team'].transform(lambda x: x.fillna(x.median()))
df_combined['distance_travel_opp'] = df_combined.groupby('opponent')['distance_travel_opp'].transform(lambda x: x.fillna(x.median()))
```

```python
# Temperature
df_combined['month'] = pd.to_datetime(df_combined['date']).dt.month

df_combined['temperature_venue'] = df_combined.groupby(['venue_country', 'month'])['temperature_venue'].transform(lambda x: x.fillna(x.median()))
df_combined['temperature_venue'] = df_combined.groupby('venue_country')['temperature_venue'].transform(lambda x: x.fillna(x.median()))
df_combined.drop(columns=['month'], inplace=True)
```

```python
# Fallback Imputation
for col in env_cols:
    df_combined[col] = df_combined[col].fillna(df_combined[col].median())
```
