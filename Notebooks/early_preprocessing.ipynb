### Calculate the *_last(n) feature at least 1 match

```python
# Making match_key to track each match
pair_key = df_combined[['team', 'opponent']].apply(
    lambda x: '||'.join(sorted([x['team'], x['opponent']])), axis=1
)

df_combined['match_key'] = df_combined['date'].astype(str) + '||' + pair_key

matches = df_combined.drop_duplicates('match_key').copy()
matches['match_key']
```

```python
# Separating team parts and opponent parts from df_combined
team_part = matches[['match_key', 'date', 'team', 'opponent', 'team_goals', 'opp_goals']].copy()
team_part = team_part.rename(
    columns={
        'team': 'team_name',
        'opponent': 'opp_name',
        'team_goals': 'goals_for',
        'opp_goals': 'goals_against',
    }
)

opp_part = matches[['match_key', 'date', 'team', 'opponent', 'team_goals', 'opp_goals']].copy()
opp_part = opp_part.rename(
    columns={
        'opponent': 'team_name',
        'team': 'opp_name',
        'opp_goals': 'goals_for',
        'team_goals': 'goals_against'
    }
)

parts = pd.concat([team_part, opp_part], ignore_index=True)
parts.head()
```

```python
# Defining each points team by goals_for (team) and goals_against (opp)
parts['points'] = np.select(
    [
        parts['goals_for'] > parts['goals_against'],    # +3 pt if win
        parts['goals_for'] == parts['goals_against']    # +1 pt if draw
    ], [3, 1], default=0                                # +0 pt if lose
)

# Making some new component according goals_for and goals_against
parts['win'] = (parts['goals_for'] > parts['goals_against']).astype(int)
parts['lose'] = (parts['goals_for'] < parts['goals_against']).astype(int)
parts['draw'] = (parts['goals_for'] == parts['goals_against']).astype(int)
parts['gd'] = parts['goals_for'] - parts['goals_against']

parts.head()
```

```python
# Sorting each match
parts = parts.sort_values(['team_name', 'date', 'match_key'])
parts.head()
```

```python
# Calculate the newest history performance columns properly
# e.g. calculate the last 4 match or below if some team just played that matches
g = parts.groupby('team_name', group_keys=False)

for i in [5, 10]:
    parts[f'points_last{i}'] = g['points'].apply(lambda x: x.shift(1).rolling(i, 1).sum())
    parts[f'gd_last{i}'] = g['gd'].apply(lambda x: x.shift(1).rolling(i, 1).sum())
    parts[f'avg_goals_last{i}'] = g['goals_for'].apply(lambda x: x.shift(1).rolling(i, 1).mean())
    parts[f'avg_conceded_last{i}'] = g['goals_against'].apply(lambda x: x.shift(1).rolling(i, 1).mean())
    parts[f'win_rate_last{i}'] = g['win'].apply(lambda x: x.shift(1).rolling(i, 1).mean())
```

```python
parts = parts.sort_values(['team_name', 'opp_name', 'date', 'match_key'])

# Group the team_name and opp_name columns to get the h2h_* columns
def compute_h2h(group):
    group = group.copy()
    
    group['h2h_points_last5'] = (
        group['points']
        .shift(1)
        .rolling(5, min_periods=1)
        .sum()
    )
    
    group['h2h_gd_last5'] = (
        group['gd']
        .shift(1)
        .rolling(5, min_periods=1)
        .sum()
    )
    
    return group

parts = parts.groupby(['team_name', 'opp_name'], group_keys=False).apply(compute_h2h)
```

```python
# Separate again the team parts and oponnent parts
feats = parts[[
    'match_key', 'team_name',
    'points_last5', 'points_last10',
    'gd_last5', 'avg_goals_last5',
    'avg_conceded_last5', 'win_rate_last10',
    'h2h_points_last5', 'h2h_gd_last5'
]]

team_feats = feats.rename(
    columns={
        'team_name': 'team',
        'points_last5': 'team_points_last5',
        'points_last10': 'team_points_last10',
        'gd_last5': 'team_gd_last5',
        'avg_goals_last5': 'team_avg_goals_last5',
        'avg_conceded_last5': 'team_avg_conceded_last5',
        'win_rate_last10': 'team_win_rate_last10',
        'h2h_points_last5': 'h2h_points_last5',
        'h2h_gd_last5': 'h2h_gd_last5'
    }
)

feats = feats.drop(columns=['h2h_points_last5', 'h2h_gd_last5'])

opp_feats = feats.rename(
    columns={
        'team_name': 'opponent',
        'points_last5': 'opp_points_last5',
        'points_last10': 'opp_points_last10',
        'gd_last5': 'opp_gd_last5',
        'avg_goals_last5': 'opp_avg_goals_last5',
        'avg_conceded_last5': 'opp_avg_conceded_last5',
        'win_rate_last10': 'opp_win_rate_last10',
        'points_last5_diff': 'points_last5_diff',
        'gd_last5_diff': 'gd_last5_diff'
    }
)

cols_old = [col for col in df_combined.columns if any(k in col for k in [
    'points_last', 'gd_last', 'avg_', 'win_rate'
])]

df_combined = df_combined.drop(
    columns=cols_old + [
        'points_last5_diff',
        'gd_last5_diff',
        'h2h_points_last5',
        'h2h_gd_last5'
    ],
    errors='ignore'
)

# Merge back the parts to its key
df_combined = df_combined.merge(team_feats, on=['match_key', 'team'], how='left')
df_combined = df_combined.merge(opp_feats, on=['match_key', 'opponent'], how='left')
```

```python
# Height Of Match Venue
df_combined['altitude_venue'] = df_combined.groupby('venue_country')['altitude_venue'].transform(lambda x: x.fillna(x.median()))
```

```python
# Miliaage

# Let the distance is zero if match is attended in home
df_combined.loc[df_combined['is_home'] == 1, 'distance_travel_team'] = df_combined.loc[df_combined['is_home'] == 1, 'distance_travel_team'].fillna(0)
df_combined.loc[df_combined['is_home'] == 0, 'distance_travel_opp'] = df_combined.loc[df_combined['is_home'] == 0, 'distance_travel_opp'].fillna(0)

# if not home, we fill the missing value by the median of team
df_combined['distance_travel_team'] = df_combined.groupby('team')['distance_travel_team'].transform(lambda x: x.fillna(x.median()))
df_combined['distance_travel_opp'] = df_combined.groupby('opponent')['distance_travel_opp'].transform(lambda x: x.fillna(x.median()))
```

```python
# Temperature
df_combined['month'] = pd.to_datetime(df_combined['date']).dt.month

# Fill the missing value according to venue_country and month
df_combined['temperature_venue'] = df_combined.groupby(['venue_country', 'month'])['temperature_venue'].transform(lambda x: x.fillna(x.median()))

# If still missing, just get the value according to the venue_country
df_combined['temperature_venue'] = df_combined.groupby('venue_country')['temperature_venue'].transform(lambda x: x.fillna(x.median()))
df_combined.drop(columns=['month'], inplace=True)
```
